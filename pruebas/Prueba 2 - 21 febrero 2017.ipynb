```julia
##[1] (i),(ii)
function mi_factorial(N::Real)
    if N < 0 || isinteger(N) == false
        error("N es negativo o no es un entero.")
    end
    prod = 1
    for i in 2:N
        prod = prod*i
    end
    return prod
end

mi_factorial(4)
```

```julia
mi_factorial(-1)
```

```julia
mi_factorial(2.1)
```

```julia
###(iii)
function mi_factorial2(N::Real)
    if N < 0 || isinteger(N) == false
        error("N es negativo o no es un entero.")
    end
    prod = 1
    i = 1
    while i <= N && i < 1000
        prod = prod*i
        i += 1
    end
    return prod
end

mi_factorial2(6)
```

```julia
###[2] (i)
function e_con_suma(N::Int)
    suma = 1 + sum([1/mi_factorial(i) for i in 1:N]) #al parecer la función mi_factorial acepta más precisión por defecto que la interna de julia (sin usar big) 
    return suma
end

e_con_suma(40)
```

```julia
using Plots
gr()
```

```julia
function graficar_error_e(N::Real, escala = :identity)
    p = scatter()
    diferencias = []
    for i in 1:N
        push!(diferencias, abs(exp(1) - e_con_suma(i)))
        p = scatter(diferencias, xscale=escala, yscale=escala)
    end
    return p
end
graficar_error_e(20)
```

```julia
function e_nuevo(N::Real)
    n = 1
    n_mas_uno = 1
    for i in 1:N
        n += n_mas_uno
        n_mas_uno = n_mas_uno/(i+1)
    end
    return n
end

e_nuevo(200) #ya acepta mucho más, haciendo pruebas en terminal llegué a 50,000,000 al menos
```

```julia
function graficar_error_e_nuevo(N::Real, escala = :identity)
    plot = scatter()
    diferencias = []
    for i in 1:N
        push!(diferencias, abs(exp(1) - e_nuevo(i)))
        plot = scatter(diferencias, xscale=escala, yscale=escala)
    end
    return plot
end
graficar_error_e_nuevo(100)
```

[3] Otra manera de calcular $e$ es $\lim_{n \to \infty}(1 + 1/n)^n$.

Grafica la distancia entre $\exp(1)$ y $(1+1/n)^n$ para distintos $n$. ¿Qué tipo de convergencia hay conforme $n \to \infty$? Descríbelo cuando menos cualitativamente con una gráfica apropiada.

**Opcional: Calcula la tasa de convergencia de manera analítica usando el desarrollo en serie de Taylor a segundo oren de $\exp(x)$.

```julia
function diferencia_limite_e(N::Real)
    diferencias = Real[]
    for i in 1:N
        push!(diferencias,exp(1) - (1 + 1/i)^i)
    end
    return diferencias
end

scatter(diferencia_limite_e(120))
```

```julia
scatter(diferencia_limite_e(120), yscale=:log10, xscale=:log10)
```

Podemos ver que la gráfica en log, log es ligeramente cóncava hacia arriba al inicio y se va acercando a una línea recta conforme $N \rightarrow \infty$. Esta tendencia es mucho más lenta que los dos métodos anteriores.
